Problem Statement 1: Employee Performance Bonus Eligibility


```
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

maxi=max(employees.values())
top=[]
for i,j in employees.items():
  if j==maxi:
    top.append(i)
print("top employees with high performance are:",*top,"(Score)",maxi)
```

Problem Statement 2: Search Query Keyword Analysis


```
import string
query=input()
query=query.lower()
query = query.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))
words=query.split()
frequency={}
for i in words:
  if i in frequency:
    frequency[i]+=1
  else:
    frequency[i]=1
result={}
for j,count in frequency.items():
  if count>1:
    result[j]=count
print(result)
```

Problem Statement 4: Email Domain Usage Analysis

```
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
gmail=0
yahoo=0
for i in emails:
  if i.endswith("@gmail.com"):
    gmail+=1
  elif i.endswith("@yahoo.com"):
    yahoo+=1
gmail_per=(gmail/len(emails))*100
yahoo_per=(yahoo/len(emails))*100
print("gmail.com:",gmail_per)
print("yahoo.com:",yahoo_per)
```

or


```
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count = {}
for i in emails:
    domain = i.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_emails = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

Problem Statement 5: Sales Spike Detection

```
sales = [1200, 1500, 900, 2200, 1400, 3000]
z=0
for i in sales:
  z+=i
avg=z/len(sales)
threshold =avg +(avg*0.3)
print("Average :",avg)
#30% = 30/100=0.3
# average=1700.0
for index in range(len(sales)):
    if sales[index] >= threshold - 10:
        print("Day", index + 1, ":", sales[index])
```

Problem Statement 6: Duplicate User ID Detection

```
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
dupli_id={}
for i in user_ids:
  if i in dupli_id:
    dupli_id[i]+=1
  else:
    dupli_id[i]=1
for j,count in dupli_id.items():
  if count>1:
    print(j,"->",count,"times")
```
