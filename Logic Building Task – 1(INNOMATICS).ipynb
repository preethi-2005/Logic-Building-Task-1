2. Pass / Fail Analyzer

```
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
for i in marks:
  if i>=50:
    pass_count+=1
  else:
    fail_count+=1
print("full class strength",len(marks))
print("pass students count",pass_count)
print("fail students coount",fail_count)
```

3.Simple Data Cleaner

```
names = [" Alice ", "bob", " CHARLIE "]
new_cleanedlist=[]
for i in names:
  i=i.strip()
  i=i.lower()
  new_cleanedlist.append(i)
print(new_cleanedlist)

```

```
names = [" Alice ", "bob", " CHARLIE "]
new_list=[]
for i in names:
  new_list.append(i.strip().lower())
print(new_list)
```

4.Message Length Analyzer

```
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
  print(i,len(i))
  if len(i)>10:
    print("This Text is more than 10 Characters")
```

5. Error Message Detector

```
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
  if i=="ERROR":
    count+=1
print(count)
```
