Problem 1: Mobile Recharge Validation System

```python
def recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Minimum recharge is ₹50")
        elif amount not in valid_plans:
            print("Invalid plan. Choose from", valid_plans)
        else:
            print(" Recharge successful!")
            break

recharge()
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock):
    for product, qty in stock.items():
        if qty < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")

products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 8,
    "Salt": 20
}

check_inventory(products)
```

 Problem 3: Student Result Processing System

```python
def result_status(marks):
    total = 0

    for m in marks:
        total += m

    avg = total / len(marks)
    print("Average:", avg)

    if avg >= 50:
        print("Pass")
    else:
        print("Fail")

result_status([60, 45, 70, 55])
```

Problem 4: Cab Fare Estimator with Retry

```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)

    if peak == "yes":
        fare *= 1.25

    return fare

while True:
    km = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ").lower()

    total = calculate_fare(km, peak)
    print("Total Fare: ₹", total)

    retry = input("Calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

Problem 5: Attendance Eligibility Checker

```python
def attendance_status(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percent = (present / len(attendance)) * 100
    print("Attendance %:", percent)

    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

record = ["P","P","A","P","P","A","P"]
print(attendance_status(record))
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        print("Strong Password ")
    else:
        print("Weak Password ")

check_password("Passs@123")
```
