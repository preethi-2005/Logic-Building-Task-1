Problem 1: Unique Words in a Sentence

```python
s=input("Enter The Sentence:\n")
k=s.split()
unique=set(k)
print("length of unique words:",len(unique))
print(unique)
```

Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
max_salary=0
max_sal_name=""
for name,salary in employees.items():
    if salary>max_salary:
        max_salary=salary
        max_sal_name=name
print("Highest Salary Getting Employee:\n",max_sal_name,max_salary)
```

Problem 3: Find Maximum and Minimum Values

```python
li=list(map(int,input().split()))
minimum=li[0]
maximum=li[0]
for i in li:
    if i>maximum:
        maximum=i
    if i<minimum:
        minimum=i
print("list eleents:",li)
print("maximum element:",maximum)
print("minimum element:",minimum)
```

Problem 4: Count Products Above a Price Threshold

```python
li=list(map(int,input().split()))
count=0
for i in li:
    if i>1000:
        print(i)
        count+=1
print("count prices greater than 1000:",count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
count=0
total=len(attendance)
for i in attendance:
    if i=='P':
        count+=1
i=(count/total)*100
print(i)
```

Problem 7: Count Character Frequency

```python
str=input()
di={}
for ch in str:
    if ch in di:
        di[ch]+=1
    else:
        di[ch]=1
print(di)
```
